```csharp
class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    public virtual void Speak()
    {
        Console.WriteLine($"{Name} издает звук.");
    }

    public virtual void Move()
    {
        Console.WriteLine($"{Name} двигается.");
    }
}

class Mammal : Animal
{
    public bool HasFur { get; set; }

    public Mammal(string name, int age, bool hasFur) : base(name, age)
    {
        HasFur = hasFur;
    }

    public virtual void FeedMilk()
    {
        Console.WriteLine($"{Name} кормит детенышей молоком.");
    }
}

class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void Move()
    {
        Console.WriteLine($"{Name} летает.");
    }

    public virtual void LayEggs()
    {
        Console.WriteLine($"{Name} откладывает яйца.");
    }
}

class Dog : Mammal
{
    public string Breed { get; set; }

    public Dog(string name, int age, bool hasFur, string breed) : base(name, age, hasFur)
    {
        Breed = breed;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} лает.");
    }
}

class Cat : Mammal
{
    public string Color { get; set; }

    public Cat(string name, int age, bool hasFur, string color) : base(name, age, hasFur)
    {
        Color = color;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} мяукает.");
    }
}

class Parrot : Bird
{
    public string SpeechAbility { get; set; }

    public Parrot(string name, int age, double wingSpan, string speechAbility) : base(name, age, wingSpan)
    {
        SpeechAbility = speechAbility;
    }

    public override void Speak()
    {
        Console.WriteLine($"{Name} может говорить: {SpeechAbility}");
    }
}

var dog = new Dog("Бобик", 5, true, "Овчарка");
var cat = new Cat("Мурка", 3, true, "Черная");
var parrot = new Parrot("Кеша", 2, 1.2, "Привет");

dog.Speak();
cat.Speak();
parrot.Speak();

dog.Move();
cat.Move();
parrot.Move();

dog.FeedMilk();
cat.FeedMilk();
parrot.LayEggs();
```
